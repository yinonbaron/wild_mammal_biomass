```python
import pandas as pd
import numpy as np
```

```python
range_data = pd.read_csv('range_data.csv')
mass_data = pd.read_csv('PanTHERIA_1-0_WR05_Aug2008.csv', sep='\t')
range_data = range_data
range_data
```

```python
merged_df = mass_data.merge(range_data, how='inner', left_on='MSW05_Binomial', right_on='binomial') 
merged_df
```

```python
merged_df['Log log total num'] = shai_func(merged_df['5-1_AdultBodyMass_g'])*merged_df['Range_size']
merged_df['Shai total num'] = shai_real_func(merged_df['5-1_AdultBodyMass_g'],merged_df['Range_size'])
```

```python
merged_df['Log log trophic num'] = shai_carn_func(merged_df['5-1_AdultBodyMass_g'])*merged_df['Range_size']
merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'Log log trophic num'] = shai_herb_func(merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'5-1_AdultBodyMass_g'])*merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'Range_size']
merged_df.loc[merged_df['6-2_TrophicLevel'] == 2,'Log log trophic num'] = shai_omni_func(merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'5-1_AdultBodyMass_g'])*merged_df.loc[merged_df['6-2_TrophicLevel'] == 2,'Range_size']


merged_df['Shai total trophic num'] = shai_carn_real_func(merged_df['5-1_AdultBodyMass_g'],merged_df['Range_size'])
merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'Shai total trophic num'] = shai_herb_real_func(merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'5-1_AdultBodyMass_g'],merged_df.loc[merged_df['6-2_TrophicLevel'] == 1,'Range_size'])
merged_df.loc[merged_df['6-2_TrophicLevel'] == 2,'Shai total trophic num'] = shai_omni_real_func(merged_df.loc[merged_df['6-2_TrophicLevel'] == 2,'5-1_AdultBodyMass_g'],merged_df.loc[merged_df['6-2_TrophicLevel'] == 2,'Range_size'])
```

```python
merged_df['Log log total mass'] = merged_df['Log log total num']*merged_df['5-1_AdultBodyMass_g']
merged_df['Shai total mass'] = merged_df['Shai total num']*merged_df['5-1_AdultBodyMass_g']

merged_df['Log log trophic mass'] = merged_df['Log log trophic num']*merged_df['5-1_AdultBodyMass_g']
merged_df['Shai total trophic mass'] = merged_df['Shai total trophic num']*merged_df['5-1_AdultBodyMass_g']
```
