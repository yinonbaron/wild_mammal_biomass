```python
import pandas as pd
import numpy as np
from sklearn import linear_model, metrics
```

```python
data = pd.read_csv('geb12617-sup-0001-suppinfo1 (1).csv')
mammals = data[(data['Taxonomic group'] =='Mammals')]
mammals
```

```python
def create_regr(data):
    lm = linear_model.LinearRegression()
    x_train = data[['Log10 body mass','Log10 study area']]
    y_train = data['Log10 population density'] + data['Log10 study area']
    lm.fit(x_train,y_train)
    print(r'z = {} {}*x {}*y'.format(lm.intercept_,lm.coef_[0],lm.coef_[1]))
    print('R^2 = {}'.format(metrics.r2_score(y_train,lm.predict(x_train))))
```

```python
def create_regr_log(data):
    lm2 = linear_model.LinearRegression()
    x_train = data['Log10 body mass']
    y_train = data['Log10 population density']
    lm2.fit(x_train.values[:,np.newaxis],y_train.values[:,np.newaxis])
    print(r'z = {} {}*x'.format(lm2.intercept_,lm2.coef_))
    print('R^2 = {}'.format(metrics.r2_score(y_train,lm2.predict(x_train.values[:,np.newaxis]))))
```

```python
create_regr(mammals)
create_regr_log(mammals)
```

```python
carnivores = data[(data['Taxonomic group'] =='Mammals') & (data['Diet (Proxy for Trophic level)'] =='Carnivore')]
herb = data[(data['Taxonomic group'] =='Mammals') & (data['Diet (Proxy for Trophic level)'] =='Herbivore')]
omni = data[(data['Taxonomic group'] =='Mammals') & (data['Diet (Proxy for Trophic level)'] =='Omnivore')]
```

```python
create_regr(carnivores)
create_regr(herb)
create_regr(omni)
```

```python
create_regr_log(carnivores)
create_regr_log(herb)
create_regr_log(omni)
```

```python
pantheria = pd.read_csv('PanTHERIA_1-0_WR05_Aug2008.csv',sep='\t')
pantheria['Log10 body mass'] = np.log10(pantheria['5-1_AdultBodyMass_g'])
pantheria['Log10 population density'] = np.log10(pantheria['21-1_PopulationDensity_n/km2']/1e2)
create_regr_log(pantheria[['Log10 body mass','Log10 population density']].dropna())
```
